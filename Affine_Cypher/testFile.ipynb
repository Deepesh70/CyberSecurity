```python
import math 
def Affine_cypher_encrypt(plain_text, key1, key2):
    result = []
   

    for char in plain_text.upper():
        if char.isalpha():
            x = ord(char) - ord('A')
            encrypted_value = ((key1 * x) + key2) % 26
            result.append(chr(encrypted_value + ord('A')))
        else:
            result.append(char)
    return "".join(result)


text = "hi i am deepesh"
key1 = input("Enter the key 1 for multiplication: ")
key2 = input("Enter the key 2 for addition: ")
if math.gcd(int(key1),26) != 1:
    raise ValueError("Key 1 and 26 are not coprime. Please enter a valid key 1.")
    exit()
else:
    result = Affine_cypher_encrypt(text, int(key1), int(key2))
    print("Encrypted Text: ", result)
```
