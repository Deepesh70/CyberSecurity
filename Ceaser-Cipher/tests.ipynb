```python
def ceaser_encryption(text, key):
    cipher = ""
    key = key % 255
    for char in text: 
        cipher = cipher + chr(ord(char) + key)
    return cipher

text = open('plaintext.txt', 'r').read()
key = 3
ciphertext = ceaser_encryption(text, key)
print("Ciphertext:", ciphertext)
# Write ciphertext to file
with open('ciphertext.txt', 'w') as f:
    f.write(ciphertext)
```

***Decryption***

```python
def CeaserDecryption(ciphertext, key):
    plaintext = ""
    for char in ciphertext:
        shifted = ord(char) - key
        shifted_char = chr(shifted)
        plaintext = plaintext + shifted_char
    return plaintext

ciphertext = open('ciphertext.txt', 'r').read()
key = 3
plaintext = CeaserDecryption(ciphertext, key)
print("Plaintext:", plaintext)
        
```
