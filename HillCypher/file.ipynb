```python
import numpy as np

def key_matrix(key):
    l = len(key)
    n = int(l**0.5)
    # print(n)

    ascii_value = []
    for i in range(l):
        characters = (ord(key[i]) - ord('A'))
        # print(characters)
        ascii_value.append(characters)

    matrix = np.zeros((n,n), dtype=int)
    # print("inside key_matrix:" , type(matrix))
    for row in range(n):
        for col in range(n):
            matrix[row,col] = ascii_value[row*n + col]
    return matrix, n
            

def text_matrix(text, n, matrix):
    ascii_values = [(ord(c) - ord('A')) for c in text]

    # Pad with 'X' if needed
    # while len(ascii_values) % n != 0:
    #     ascii_values.append(ord('X') - ord('A'))

    cipher_text = ""

    for i in range(0, len(ascii_values), n):
        pairs = ascii_values[i:i+n]
        # print("pair:", pairs)

        p_matrix = np.array(pairs).reshape(n, 1)

        product = np.dot(matrix, p_matrix) % 26
        # print("product:\n", product)

        cipher_chars = [
            chr(int(val) + ord('A')) for val in product.flatten()
        ]

        cipher_text = cipher_text + "".join(cipher_chars)

    return cipher_text

def adjugate_matrix(matrix):
    det = int(round(np.linalg.det(matrix)))
    inv = np.linalg.inv(matrix)
    adj = np.round(det * inv).astype(int)
    return adj


def inverse_matrix(k_matrix):
    determinant = int(round(np.linalg.det(k_matrix))) % 26
    for i in range(1, 26):
        if ( determinant * i ) % 26 == 1:
            determinant_inverse = i
            break
    adj = adjugate_matrix(k_matrix)
    inverse_key = (determinant_inverse * adj) % 26
    return inverse_key



def decryption(cipher_text, inverse_key_matrix, n):
    ascii_value = [ord(c) - ord('A') for c in cipher_text]
    decrypt_text = ""

    for i in range(0, len(ascii_value), n):
        block = ascii_value[i:i+n]
        pair_matrix = np.array(block).reshape(n, 1)

        product = np.dot(inverse_key_matrix, pair_matrix) % 26

        decrypt_chars = [
            chr(int(val) + ord('A')) for val in product.flatten()
        ]

        decrypt_text += "".join(decrypt_chars)

    return decrypt_text


if __name__ == "__main__":
    choice = input("Enter 1 for encryption and 2 for decryption 0 for exit: ")
    if choice == "1":
        with open("./plaintext.txt", "r") as file:
            plaintext = file.read().upper().replace(" ", "")
        key = input("Enter the key form a square matrix: ")
        k_matrix, n = key_matrix(key)
        # print(type(k_matrix))
        # print("Key matrix:\n", k_matrix)
        determinant = np.linalg.det(k_matrix)
        print("Determinant of the key matrix:", determinant)

        if determinant == 0:
            print("Key matrix is not invertible. Please provide a valid key, Inverse should exits.")
            exit()

        cipher = text_matrix(plaintext, n, k_matrix)
        with open("ciphertext.txt", "w") as file:
            file.write(cipher)

    if choice == "2":
        with open("./ciphertext.txt", "r") as file:
            cipher = file.read().upper().replace(" ", "")
        key = input("Enter the key form a square matrix: ")
        k_matrix, n = key_matrix(key)
        invere_key_matrix = inverse_matrix(k_matrix)
        decrypt_text = decryption(cipher, invere_key_matrix, n)
        print("Decrypted text:", decrypt_text)
        with open("decrypted_text.txt", "w") as file:
            file.write(decrypt_text)
    if choice == "0":
        exit()
```

```python
import numpy as np
import re

def key_matrix(key):
    key = re.sub(r'[^A-Z]', '', key.upper())
    l = len(key)
    n = int(l**0.5)
    if n * n != l:
        raise ValueError("Key length must be a perfect square (e.g., 4, 9, 16).")

    ascii_value = [ord(c) - ord('A') for c in key]
    matrix = np.array(ascii_value).reshape(n, n)
    return matrix, n

def text_matrix(text, n, matrix):

    clean_text = re.sub(r'[^A-Z]', '', text.upper())
    ascii_values = [ord(c) - ord('A') for c in clean_text]

    while len(ascii_values) % n != 0:
        ascii_values.append(ord('X') - ord('A'))

    cipher_text = ""
    for i in range(0, len(ascii_values), n):
        block = np.array(ascii_values[i:i+n]).reshape(n, 1)
        product = np.dot(matrix, block) % 26
        cipher_chars = [chr(int(val) + ord('A')) for val in product.flatten()]
        cipher_text += "".join(cipher_chars)
    return cipher_text

def get_mod_inverse(a, m):
    for x in range(1, m):
        if (a * x) % m == 1:
            return x
    return None

def inverse_matrix(k_matrix, n):
    det = int(round(np.linalg.det(k_matrix)))
    det_mod26 = det % 26
    
    det_inv = get_mod_inverse(det_mod26, 26)
    if det_inv is None:
        raise ValueError(f"Determinant {det_mod26} has no inverse mod 26. Use a different key.")

    adjugate = np.round(det * np.linalg.inv(k_matrix)).astype(int)
    inverse_key = (det_inv * adjugate) % 26
    return inverse_key

def decryption(cipher_text, inverse_key_matrix, n):

    ascii_values = [ord(c) - ord('A') for c in cipher_text]
    decrypt_text = ""
    for i in range(0, len(ascii_values), n):
        block = np.array(ascii_values[i:i+n]).reshape(n, 1)
        product = np.dot(inverse_key_matrix, block) % 26
        decrypt_chars = [chr(int(val) + ord('A')) for val in product.flatten()]
        decrypt_text += "".join(decrypt_chars)
    return decrypt_text

if __name__ == "__main__":
    choice = input("1: Encrypt, 2: Decrypt, 0: Exit: ")
    
    if choice == "1":
        key_input = input("Enter key (square length): ")
        try:
            k_matrix, n = key_matrix(key_input)
            
            with open("./plaintext.txt", "r") as f:
                content = f.read()
            
            cipher = text_matrix(content, n, k_matrix)
            with open("ciphertext.txt", "w") as f:
                f.write(cipher)
            print("Encryption successful. Check ciphertext.txt")
        except Exception as e:
            print(f"Error: {e}")

    elif choice == "2":
        key_input = input("Enter the same key: ")
        try:
            k_matrix, n = key_matrix(key_input)
            inv_matrix = inverse_matrix(k_matrix, n)
            
            with open("ciphertext.txt", "r") as f:
                cipher = f.read()
                
            decoded = decryption(cipher, inv_matrix, n)
            print(f"Decrypted: {decoded[:100]}...") 
            with open("decrypted_text.txt", "w") as f:
                f.write(decoded)
        except Exception as e:
            print(f"Error: {e}")
```
