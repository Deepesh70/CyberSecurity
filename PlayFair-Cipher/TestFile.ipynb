```python
def find_position(matrix, char):
    for r, row in enumerate(matrix):
        if char in row:
            return r, row.index(char)
    return None



def playfair_encrypt(matrix, plaintext):
    text = plaintext.upper().replace(" ", "").replace("J","I")
    prepared_text = ""
    i = 0
    while( i< len(text)):
        a = text[i]
        if (i+1) < len(text):
            b = text[i+1]
        else:
            b = 'X'
        # print(f"Pairing letters: {a}, {b}")
        if a==b:
            prepared_text = prepared_text + a + 'X'
            i = i + 1
            # print("when a == b ", prepared_text)
        else:
            prepared_text = prepared_text + a + b
            i = i +2
            # print("when a != b ", prepared_text)
        # print("outside the if else condition: ",prepared_text)

        ciphertext = ""
        for i in range(0, len(prepared_text) , 2):
            row1, col1 = find_position(matrix, prepared_text[i])
            row2, col2 = find_position(matrix , prepared_text[i+1])
            # print(f"Positions: {prepared_text[i]}({row1},{col1}), {prepared_text[i+1]}({row2},{col2})")
            if row1 == row2:
                ciphertext= ciphertext + matrix[row1][(col1 + 1) % 5]
                ciphertext  = ciphertext + matrix[row1][(col2 + 1) % 5]
                # print("Same Row Ciphertext: ",ciphertext)
            elif col1 == col2:
                ciphertext = ciphertext + matrix[(row1 + 1) %5][col1]
                ciphertext = ciphertext + matrix[(row2 + 1) %5][col1]
                # print("Same Column Ciphertext: ",ciphertext)
            else:
                ciphertext = ciphertext + matrix[row1][col2]
                ciphertext = ciphertext + matrix[row2][col1]
                # print("Rectangle Rule Ciphertext: ",ciphertext)
    return ciphertext
matrix = [
    ['K', 'E', 'Y', 'W', 'O'],
    ['R', 'D', 'A', 'B', 'C'],
    ['F', 'G', 'H', 'I', 'L'],
    ['M', 'N', 'P', 'Q', 'S'],
    ['T', 'U', 'V', 'X', 'Z']
]

encryptied_text = playfair_encrypt(matrix, 'HELLO')
print("Encrypted Text:", encryptied_text)
```
