PlayFair Encryption Algorithm

```python
from string import punctuation

def create_matrix(key):
    key = key.upper()
    alphabets = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    digit = "1234567890"
   
    # print("Modified Key: ", key)
    
    combined = []
    for char in key:
        if char not in combined and char not in punctuation:
            combined.append(char)
    
    # print("After adding key: ", combined)
    
    for char in alphabets:
        if char not in combined:
            combined.append(char)
    for char in digit:
        if char not in combined:
            combined.append(char)
    # print("After adding alphabets and digits: ", combined)
    
    # print("Length of combined: ", len(combined))
    matrix = []
    for i in range(0,6):
        row= []
        for j in range(0,6):
            row.append(combined[6*i + j])
        matrix.append(row)
    # print("Matrix: ", matrix)
    return matrix   

# def print_matrix(matrix):
#     print("PlayFair Matrix:")
#     for i in range(0,6):
#         print(matrix[i])

def EncryptionPlayFair(matrix, plaintext): 
    ciphertext = ""
    i = 0
    plaintext = plaintext.upper() 

    while i < len(plaintext):
        char1 = plaintext[i]
        if i + 1 < len(plaintext):
            char2 = plaintext[i+1]
            if char1 == char2:
                char2 = 'X'
                i = i + 1
            else:
                i = i + 2
        else:
            char2 = 'X'
            i = i + 1

        row1, col1, row2, col2 = -1, -1, -1, -1
        for r in range(6):
            for c in range(6):
                if matrix[r][c] == char1:
                    row1, col1 = r, c
                if matrix[r][c] == char2:
                    row2, col2 = r, c


        if row1 == row2:
            ciphertext = ciphertext + matrix[row1][(col1 + 1) % 6]
            ciphertext = ciphertext + matrix[row2][(col2 + 1) % 6]
        elif col1 == col2:
            ciphertext = ciphertext + matrix[(row1 + 1) % 6][col1]
            ciphertext = ciphertext + matrix[(row2 + 1) % 6][col2]
        else:
            ciphertext = ciphertext + matrix[row1][col2]
            ciphertext = ciphertext + matrix[row2][col1]
            
    return ciphertext

if __name__ == "__main__":
    key = input("Enter the Key: ")
    matrix = create_matrix(key)
    # print_matrix(matrix)
    
    with open("./plaintext.txt", "r") as file:
        plain_text = file.read()
    
    plaintext = "".join([c.upper() for c in plain_text if c.isalnum()])
    
    print("Modified Plain Text:", plaintext)
    encrypted = EncryptionPlayFair(matrix, plaintext)
    print("Encrypted Text:", encrypted)



if __name__ == "__main__":
    key = input("Enter the Key: ")
    matrix = create_matrix(key)
    # print_matrix(matrix)
    with open("./plaintext.txt", "r") as file:
        plaintext = file.read()
        
    for char in plaintext:
        if char in punctuation or char == ' ' or char == '\n':
            plaintext = plaintext.replace(char, '')
    # print("Modified Plain Text:", plaintext)
    encrypted = EncryptionPlayFair(matrix, plaintext)
    # print("Encrypted Text:", encrypted)
    with open("./ciphertext.txt", "w") as file:
        file.write(encrypted)
```

```python
from string import punctuation

def create_matrix(key):
    key = key.upper()
    alphabets = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    digit = "1234567890"
   
    # print("Modified Key: ", key)
    
    combined = []
    for char in key:
        if char not in combined and char not in punctuation:
            combined.append(char)
    
    # print("After adding key: ", combined)
    
    for char in alphabets:
        if char not in combined:
            combined.append(char)
    for char in digit:
        if char not in combined:
            combined.append(char)
    # print("After adding alphabets and digits: ", combined)
    
    # print("Length of combined: ", len(combined))
    matrix = []
    for i in range(0,6):
        row= []
        for j in range(0,6):
            row.append(combined[6*i + j])
        matrix.append(row)
    # print("Matrix: ", matrix)
    return matrix 

def decryptionPlayFair(matrix, ciphertext):
    decrypted = ""
    ciphertext = "".join([c.upper() for c in ciphertext if c.isalnum()])
    i = 0
    while i < len(ciphertext):
        char1 = ciphertext[i]
        if i + 1 < len(ciphertext):
            char2 = ciphertext[i+1]
            if char1 == char2:
                char2 = 'X'
                i = i+ 1
            else:
                i = i + 2
        else:
            char2 = 'X'
            i = i + 1
        row1 , col1, row2, col2 = -1, -1, -1, -1
        for r in range(6):
            for c in range(6):
                if matrix[r][c] == char1:
                    row1, col1 = r,c 
                if matrix[r][c] == char2:
                    row2, col2 = r,c
        if row1 == row2:
            decrypted  = decrypted + matrix[row1][(col1 - 1) % 6]
            decrypted  = decrypted + matrix[row2][(col2 - 1) % 6]
        elif col1 == col2:
            decrypted  = decrypted + matrix[(row1 - 1) % 6][col1]
            decrypted  = decrypted + matrix[(row2 - 1) % 6][col2]
        else:
            decrypted  = decrypted + matrix[row1][col2]
            decrypted  = decrypted + matrix[row2][col1]
    return decrypted
    

if __name__ == "__main__":
    key = input("Enter the Key: ")
    matrix = create_matrix(key)

    with open("./ciphertext.txt", "r") as file:
        ciphertext = file.read()

    decrypted = decryptionPlayFair(matrix, ciphertext)

    with open("./Recovered.txt", "w") as file:
        file.write(decrypted)

    print("Decrypted Text:", decrypted)
```
