Menu Driven Program for PlayFair Cipher

```python
from string import punctuation

def create_matrix(key):
    key = key.upper()
    alphabets = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    digit = "1234567890"
    
    combined = []
    for char in key:
        if char not in combined and char not in punctuation:
            combined.append(char)
    
    for char in alphabets:
        if char not in combined:
            combined.append(char)
    for char in digit:
        if char not in combined:
            combined.append(char)
    
    matrix = []
    for i in range(0, 6):
        row = []
        for j in range(0, 6):
            row.append(combined[6*i + j])
        matrix.append(row)
    return matrix

def print_matrix(matrix):
    print("\nPlayFair Matrix:")
    for row in matrix:
        print(" ".join(row))

def encrypt_playfair(matrix, plaintext):
    ciphertext = ""
    i = 0
    plaintext = plaintext.upper()

    while i < len(plaintext):
        char1 = plaintext[i]
        if i + 1 < len(plaintext):
            char2 = plaintext[i+1]
            if char1 == char2:
                char2 = 'X'
                i = i + 1
            else:
                i = i + 2
        else:
            char2 = 'X'
            i = i + 1

        row1, col1, row2, col2 = -1, -1, -1, -1
        for r in range(6):
            for c in range(6):
                if matrix[r][c] == char1:
                    row1, col1 = r, c
                if matrix[r][c] == char2:
                    row2, col2 = r, c

        if row1 == row2:
            ciphertext += matrix[row1][(col1 + 1) % 6] + matrix[row2][(col2 + 1) % 6]
        elif col1 == col2:
            ciphertext += matrix[(row1 + 1) % 6][col1] + matrix[(row2 + 1) % 6][col2]
        else:
            ciphertext += matrix[row1][col2] + matrix[row2][col1]
            
    return ciphertext

def decrypt_playfair(matrix, ciphertext):
    decrypted = ""
    ciphertext = "".join([c.upper() for c in ciphertext if c.isalnum()])
    
    for i in range(0, len(ciphertext), 2):
        char1 = ciphertext[i]
        char2 = ciphertext[i+1]
        row1, col1, row2, col2 = -1, -1, -1, -1
        
        for r in range(6):
            for c in range(6):
                if matrix[r][c] == char1:
                    row1, col1 = r, c
                if matrix[r][c] == char2:
                    row2, col2 = r, c
        
        if row1 == row2:
            decrypted += matrix[row1][(col1 - 1) % 6] + matrix[row2][(col2 - 1) % 6]
        elif col1 == col2:
            decrypted += matrix[(row1 - 1) % 6][col1] + matrix[(row2 - 1) % 6][col2]
        else:
            decrypted += matrix[row1][col2] + matrix[row2][col1]
    
    return decrypted

def main():
    print("\nPLAYFAIR CIPHER\n")
    
    while True:
        print("\n1. Encrypt")
        print("2. Decrypt")
        print("3. Show Matrix")
        print("4. Exit")
        
        choice = input("\nChoice: ")
        
        if choice == '1':
            key = input("Key: ")
            matrix = create_matrix(key)
            text = input("Text: ")
            text = "".join([c.upper() for c in text if c.isalnum()])
            result = encrypt_playfair(matrix, text)
            print(f"\nEncrypted: {result}")
        
        elif choice == '2':
            key = input("Key: ")
            matrix = create_matrix(key)
            text = input("Ciphertext: ")
            result = decrypt_playfair(matrix, text)
            print(f"\nDecrypted: {result}")
        
        elif choice == '3':
            key = input("Key: ")
            matrix = create_matrix(key)
            print_matrix(matrix)
        
        elif choice == '4':
            break
        
        else:
            print("\nInvalid choice!")

if __name__ == "__main__":
    main()
```
