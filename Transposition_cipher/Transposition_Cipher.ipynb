```python
def matrix_formation(plain_text, key_length):
    space_length = (key_length - (len(plain_text) % key_length)) % key_length
    plain_text = plain_text+ "X" * space_length
    matrix = []
    for i in range(0, len(plain_text), key_length):
        row = list(plain_text[i:i+key_length])
        # print(row)    
        matrix.append(row)
    return matrix
```

```python

def transpose(matrix, key_indices):
    transposed = []
    for i in key_indices:

        column = []   # create empty list for this column

        for row in matrix:

            if i < len(row):
                column.append(row[i])

        # Add this column to transposed list
        transposed.append(column)

    return transposed
```

```python
matrix = matrix_formation(plain_text_noSpace, key_length)
print(matrix)
key_indices = [int(char) - 1 for char in str(key)]
print(key_indices)
matrix_transpose = transpose(matrix, key_indices)
print(matrix_transpose)
```

```python
def Encryption_transposition(matrix_transpose):
    cipher_text= ""
    for c in matrix_transpose:
        cipher_text = cipher_text + ''.join(c)
    return cipher_text
```

```python
cipher = Encryption_transposition(matrix_transpose)
print(cipher)
```
